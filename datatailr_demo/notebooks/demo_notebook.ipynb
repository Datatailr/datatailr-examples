```python
from datetime import date, datetime
import numpy as np
import pandas as pd
from perspective.widget import PerspectiveWidget

get_ipython().run_line_magic('load_ext', 'datatailr.scheduler.jupyter_magic')

N = 10_000
data = pd.DataFrame({
    "int": np.arange(N),
    "float": [i * 1.5 for i in range(N)],
    "bool": [True for i in range(N)],
    "date": [date.today() for i in range(N)],
    "datetime": [datetime.now() for i in range(N)],
    "string": [str(i) for i in range(N)]
})
```

## Perspective Widget

This is a demo of the Perspective Widget. Using the [perspective library](https://perspective-dev.github.io/), we can create a widget that displays a table of data or a chart.

Perspective is an interactive analytics and data visualization component, which is especially well-suited for large and/or streaming datasets.

Use it to create user-configurable reports, dashboards, notebooks and applications, then deploy stand-alone in the browser, or in concert with Python and/or Jupyterlab.

```python
PerspectiveWidget(data, aggregates={"datetime": "any"}, sort=[["date", "desc"]])
```

## Interactive Dashboard

```python
import numpy as np
from bqplot import pyplot as plt
import bqplot
import ipywidgets as widgets


n = 200

x = np.linspace(0.0, 10.0, n)
y = np.cumsum(np.random.randn(n)*10).astype(int)

label_selected = widgets.Label(value="Selected: 0")
fig = plt.figure( title='Histogram')
np.random.seed(0)
hist = plt.hist(y, bins=25)
hist.scales['sample'].min = float(y.min())
hist.scales['sample'].max = float(y.max())
display(fig)
fig.layout.width = 'auto'
fig.layout.height = 'auto'
fig.layout.min_height = '300px' # so it shows nicely in the notebook
fig.layout.flex = '1'
```

```python
fig = plt.figure( title='Line Chart')
np.random.seed(0)
n = 200
p = plt.plot(x, y)
fig
```

```python
fig.layout.width = 'auto'
fig.layout.height = 'auto'
fig.layout.min_height = '300px' # so it shows nicely in the notebook
fig.layout.flex = '1'
```

```python
brushintsel = bqplot.interacts.BrushIntervalSelector(scale=p.scales['x'])
```

```python
def update_range(*args):
    label_selected.value = "Selected range {}".format(brushintsel.selected)
    mask = (x > brushintsel.selected[0]) & (x < brushintsel.selected[1])
    hist.sample = y[mask]
    
brushintsel.observe(update_range, 'selected')
fig.interaction = brushintsel
```
