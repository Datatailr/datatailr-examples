```python
from datetime import date, datetime
import numpy as np
import pandas as pd
from perspective.widget import PerspectiveWidget

N = 10_000
data = pd.DataFrame({
    "int": np.arange(N),
    "float": [i * 1.5 for i in range(N)],
    "bool": [True for i in range(N)],
    "date": [date.today() for i in range(N)],
    "datetime": [datetime.now() for i in range(N)],
    "string": [str(i) for i in range(N)]
})
```

## Perspective Widget

This is a demo of the Perspective Widget. Using the [perspective library](https://perspective-dev.github.io/), we can create a widget that displays a table of data or a chart.

Perspective is an interactive analytics and data visualization component, which is especially well-suited for large and/or streaming datasets.

Use it to create user-configurable reports, dashboards, notebooks and applications, then deploy stand-alone in the browser, or in concert with Python and/or Jupyterlab.

```python
PerspectiveWidget(data, aggregates={"datetime": "any"}, sort=[["date", "desc"]])
```

```python
df_stream = pd.DataFrame(np.random.randn(400, 4), columns=list('ABCD')).cumsum()
PerspectiveWidget(df_stream, plugin='y_line', columns=['B', 'C', 'D'])
```
